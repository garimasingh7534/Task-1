# Distribution of Gender By using bar chart.(Categorical variable)

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Gender",data=df,hue="Gender",
              palette={"male":"steelblue","female":"salmon"},legend=False)
plt.title("Distribution of Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

# Distribution of age by usinf histogram.(Continous variable)

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Age"], bins=10, kde= True, color="skyblue")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
